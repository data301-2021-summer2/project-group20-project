```python
import seaborn as sns
import sys 
sys.path.append('..')
from scripts import project_functions as pf # Method Chain located in scripts directory
raw_data = "../../data/raw/listings.csv"
df = pf.load_and_process(raw_data) 
display(df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcParams['figure.figsize'] = [10, 10]
df_new=df.filter(['Days of Stay','Price per Night ($)','# of tenants','Total Reviews','Bayesian Rating','Room Type'])
q4_plot = sns.scatterplot(data=df_new,x='Price per Night ($)',y='Total Reviews',hue='# of tenants',size='Total Reviews',sizes=(10,200),alpha=0.35)
q4_plot.spines['right'].set_visible(False)
q4_plot.spines['top'].set_visible(False)
q4_plot.set(title='Total Reviews vs. Price Per Night ($)')
plt.legend(bbox_to_anchor=(1.3,0.8))
plt.show()
```

```python
df_new=df.filter(['Days of Stay','Price per Night ($)','# of tenants','Total Reviews','Bayesian Rating','Room Type'])
q3_plot = sns.scatterplot(data=df_new,y='Bayesian Rating',x='Total Reviews',alpha=0.5,size='Total Reviews',sizes=(0,150),hue='# of tenants')
q3_plot.set(title='Bayesian Rating vs. Total Reviews')
sns.despine()
plt.legend(bbox_to_anchor=(1.2,0.7))
plt.show()
```

```python
import pandas as pd
df_new=df.filter(['Days of Stay','Price per Night ($)','# of tenants','Total Reviews','Bayesian Rating','Room Type'])
q34_plot = sns.pairplot(df_new,corner=True,hue='# of tenants',diag_kind='kde',palette='Set2')
q34_plot.fig.suptitle("Catgorical Correlations")
```
